```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go

df = pd.read_csv("result.csv")
df.head()
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=list(range(1000)),
        y=df["N=0"],
        mode="lines",
        name="N=0",
    )
)
fig.add_trace(
    go.Scatter(
        x=list(range(1000)),
        y=df["N=5"],
        mode="lines",
        name="N=5",
    )
)
fig.add_trace(
    go.Scatter(
        x=list(range(1000)),
        y=df["N=50"],
        mode="lines",
        name="N=50",
    )
)
fig.update_layout(
    title="N=0,5,50",
    xaxis_title="Episodes",
    yaxis_title="Steps per episode",
)
fig.show()
```
